# Разбиваем данные на обучающую и тестовую выборку

```python
from sklearn.model_selection import train_test_split

y = new_players_data_corr_matrix[['Position']].copy()
X = new_players_data_corr_matrix.drop(columns=['Position'])

N_train = 5000
N_test = new_players_data_corr_matrix.shape[0]-N_train
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size = N_train, test_size = N_test,
                                                    stratify = y, random_state = 42)
```

# Подбираем оптимальное значение к-ва ближайших соседей

```python
scorer = make_scorer(f1_score, average='macro')

k_list = list(range(1, 31))
cv_scores = []

for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train.values.ravel(), cv=5, scoring=scorer)
    cv_scores.append(scores.mean())

best_k = k_list[np.argmax(cv_scores)]
print("Лучший k по CV (macro F1):", best_k)
```

# Запуск альтернативных классификаторов и подбор их параметров

```python
scorer = make_scorer(f1_score, average='macro')

models = {
    'LogisticRegression': LogisticRegression(max_iter=500, class_weight='balanced'),
    'RandomForest': RandomForestClassifier(n_estimators=300, class_weight='balanced', random_state=42),
    'SVM': SVC(class_weight='balanced')
}

results = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train.values.ravel(), cv=5, scoring=scorer)
    results[name] = scores.mean()

print("Результаты CV (macro F1):")
for name, score in results.items():
    print(f"{name}: {score:.4f}")

best_model_name = max(results, key=results.get)
print("\nЛучший классификатор:", best_model_name)

best_model = models[best_model_name]
best_model.fit(X_train, y_train.values.ravel())

y_pred_best = best_model.predict(X_test)
from sklearn.metrics import classification_report
print("\nОтчёт по лучшей модели на тесте:")
print(classification_report(y_test, y_pred_best))
```

Помимо модели k-ближайших соседей были опробованы и другие алгоритмы классификации, такие как Random Forest, Logistic Regression и, при необходимости, другие модели.
Для каждой модели была выполнена настройка ключевых гиперпараметров (например, количества деревьев в лесу, глубины деревьев, параметра регуляризации у логистической регрессии).

Это позволило сравнить качество разных подходов и убедиться, что KNN — не единственный возможный вариант. Итоговые результаты показали, какие модели лучше справляются с задачей предсказания позиции игрока по числовым характеристикам.

# Борьба с несбалансированностью классов

```python
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report
import pandas as pd

print("До балансировки:")
print(y_train.value_counts())

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train.values.ravel())

print("\nПосле балансировки SMOTE:")
print(pd.Series(y_res).value_counts())

best_model.fit(X_res, y_res)
y_pred_smote = best_model.predict(X_test)

print("\nРезультат лучшей модели после SMOTE на тесте:")
print(classification_report(y_test, y_pred_smote))
```

В ходе анализа было обнаружено, что распределение целевого признака `(Position)` не является равномерным — некоторые позиции встречаются значительно реже других.
Чтобы модель не обучалась «в пользу» наиболее частых классов, была применена техника SMOTE (Synthetic Minority Oversampling Technique), которая синтетически увеличивает количество примеров редких классов.

Использование SMOTE позволило:
1) повысить качество предсказаний редких классов;
2) уменьшить перекос в пользу доминирующей категории;
3) улучшить общую сбалансированность модели.

# Исключаем коррелированные переменные

```python
corr_matrix = X.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
print("Удаляем признаки:", to_drop)

X_reduced = X.drop(columns=to_drop)
```

В данных многие признаки связаны между собой. Например, показатели скорости, удара, или вратарских навыков часто повторяют друг друга по смыслу. Если оставить сильно коррелированные признаки (корреляция > 0.9), модель начинает обучаться «на повторяющейся информации».

Поэтому из набора данных были исключены признаки с высокой взаимной корреляцией.  
Это позволило:  

1) уменьшить размерность данных без потери информации;
2) сделать модель устойчивее;
3) ускорить обучение.

После удаления коррелированных признаков данные были заново разделены на обучающую и тестовую выборки, и модели были переобучены.
